```python
newdf=df.loc[((df['State']=="Andaman and Nicobar Islands") & (df['Crop']=='Arecanut') )]
newdf
```

```python
from sklearn import preprocessing, svm 
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression 
```

```python
X = np.array(newdf['Area']).reshape(-1, 1) 
y = np.array(newdf['Production']).reshape(-1, 1)
```

```python
newdf.isnull().sum()
```

```python
newdf.dropna()
```

```python
newdf.isnull().sum()
newdf.dropna()
newdf.isnull().sum()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)

print(y_train)
regr = LinearRegression()
regr.fit(X_train, y_train) 
print(regr.score(X_test, y_test))

result=regr.predict([[502]])
round(result[0][0])
```

```python
w0=regr.coef_[0]   # this will hold the value of the slope 
w1=regr.intercept_
```

```python
def predict(mi):
    exp=mi      #yahan apan ne min ka data diya 
    res=w1*exp+w0
    print("Expected max:: ",round(res[0],2))

#     exp=ma # yahan par apan ne max ka data diya
#     res2=(exp-w0)/w1
#     print("Expected min:: ",round(res2[0],2))
    
predict(502)
```
